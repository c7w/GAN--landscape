```python
from PIL import Image
from numpy import asarray, abs, amax
import matplotlib.pyplot as plt
import matplotlib as mpl
from pathlib import Path
```

```python

save = Path("../test_save/baseline_noise")

resize = 2

def get(idx: int):
    res = {}
    for inf in save.iterdir():
        if inf.is_dir():
            lst = list(sorted((inf.iterdir())))
            img = Image.open(lst[idx])
            res[inf.name] = img
            img.resize((int(img.width * resize), int(img.height * resize)))
    return res

def manifest(idx: int):
    all_img = get(idx)
    fig = plt.figure(figsize=(30,30))

    X = 3
    Y = 3

    for i, (k, v) in enumerate(all_img.items()):
        fig.add_subplot(X, Y, i + 1)
        plt.title(k)
        plt.axis('off')
        plt.imshow(v)

    plt.show()
```

```python
import random
manifest(random.randint(0, 999))
```
